```python
# Importación de librerías necesarias para carga de archivo
import pandas as pd
import os
```

```python
# Conversión de datos object a string
df['DATE']= df['DATE'].astype('string')
df['ITEM_CODE']= df['ITEM_CODE'].astype('string')
df['ITEM_DESCRIPTION']= df['ITEM_DESCRIPTION'].astype('string')
df['MANUFACTURER']= df['MANUFACTURER'].astype('string')
df['BRAND']= df['BRAND'].astype('string')
df['FORMAT']= df['FORMAT'].astype('string')
df['SEGMENT']= df['SEGMENT'].astype('string')
df['CATEGORY']= df['CATEGORY'].astype('string')
df['REGION']= df['REGION'].astype('string')
```

```python
# Histograma de ventas de la categoría
fig, ax= plt.subplots()
palette= sns.color_palette('PuRd', 6) 
subtono= palette[4]

sns.histplot(df_ventas, x= 'TOTAL_VALUE_SALES', color= subtono)
ax.set(title= 'Histograma de ventas FABRIC TREATMENT and SANIT', xlabel= '', ylabel= '')
sns.despine()
```

### Tendencia de ventas en el tiempo

```python
# Gráfico de línea de la tendencia de ventas
fig, ax= plt.subplots()
palette= sns.color_palette('PuRd', 6) 
subtono= palette[4]

sns.lineplot(df_ventas, x= 'DATE', y='TOTAL_VALUE_SALES', errorbar= None, color= subtono)
ax.set(title= 'Tendencia de ventas 2022 - 2023', xlabel= '', ylabel= '')
ax.xaxis.set_ticks([])
sns.despine()
```

```python
# Ventas promedio por región
df_ventas.groupby('REGION')['TOTAL_VALUE_SALES'].mean().sort_values(ascending= False)
```

```python
# Gráfico de líneas de tendencia de ventas de cada región por año 
g= sns.FacetGrid(df_ventas, col='anio', hue= 'REGION', col_wrap=2, height=5,  sharex=False, palette='Accent')
g.map_dataframe(sns.lineplot, x= 'cuatrimestre', y= 'TOTAL_VALUE_SALES', errorbar= None).add_legend()
g.fig.suptitle('Tendencia de ventas')
g.set_titles(col_template= "Dominio de las áreas 2 y 5 en {col_name}")
g.set_axis_labels('', '')

for ax in g.axes: # Resaltar líneas con mayores ventas
    for line in ax.lines:
        if line.get_label() in ('ÁREA 2', 'ÁREA 5'):
            line.set_linewidth(4)
        else:
            line.set_linewidth(2)

plt.subplots_adjust(top=0.8)
plt.xticks(rotation=45)  
g._legend.set_title('REGIÓN')
sns.despine()           
```

```python
# Se reemplazan los nombres de las regiones para simplificar la leyenda
df_ventas['REGION']=df_ventas['REGION'].replace({'TOTAL AUTOS AREA 1': 'ÁREA 1',
                                                 'TOTAL AUTOS AREA 2': 'ÁREA 2',
                                                 'TOTAL AUTOS AREA 3': 'ÁREA 3',
                                                 'TOTAL AUTOS AREA 4': 'ÁREA 4',
                                                 'TOTAL AUTOS AREA 5': 'ÁREA 5',
                                                 'TOTAL AUTOS AREA 6': 'ÁREA 6'})
```

```python
# Tendencia de ventas por segmento en todo el periodo 2022 - 2023
fig, ax= plt.subplots()
sns.lineplot(df_ventas, x= 'cuatrimestre', y= 'TOTAL_VALUE_SALES', hue= 'SEGMENT', errorbar=None, palette='Accent', linewidth= 2)
ax.set(title= 'Tendencia de ventas por segmento 2022 -2023', xlabel= 'Cuatrimestre', ylabel='')

plt.xticks(rotation=45)
sns.despine()
```

#### Observaciones: 
- El periodo que se analiza comienza en 2022 y termina en el tercer cuatrimestre de 2023.<br><br>
- La categoría que ocupa el 100 % del negocio es Fabric Treatment and Sanit.<br><br>
- El histograma de ventas indica que la frecuencia de compra más común es de productos de bajo costo, sin embargo hay una gran influencia de compras de cantidades atípicas.<br><br>
- La tendencia de ventas va en ascenso durante todo el periodo, especialmente en las áreas 2 y 5, el área 2 tiene un desarrollo importante a mediados del primer cuatrimestre de 2022 y decrece a finales del año, sin embargo asciende durante 2023.
